```python
import pandas as pd
from tensorflow import keras
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn import metrics
```

```python
#Importing the data set

df = pd.read_csv('Mobile_Dataset.csv')
```

```python
#Display data columns

df.columns
```

```python
#display the data table shape

df.shape
```

```python
#display data type of each column

df.dtypes
```

```python
#Checking for null values in each data entry

df.isnull()
```

```python
#Obtaining the total number of cells with null values in each column

df.isnull().sum()
```

```python
#Analyzing the dataset

analysis = df.describe()
print(analysis)
```

```python
#Converting data entries into values

dataset = df.values
print(dataset)
```

```python
#Imputation of missing values
#The missing values(Null values and nonnumeric values) are replaced with the mean value of the specific column 

imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(dataset)
print(dataset)
```

```python
#Discrete optimization using casting inputs

dataset = dataset.astype('float32')
print(dataset)
```

```python
#Defining input features (X) and predicting features (Y) 
x = dataset[:,0:20]
y = dataset[:,20]
```

```python
#Normalization of x data using MinMax scaler function

min_max_scaler = preprocessing.MinMaxScaler()
x = min_max_scaler.fit_transform(x)
print(x)
```

```python
#Splitting the data set into training data set and (testing and validation) data set. 80% as the training data set and 20% as the testing and validation data set. 

x_train, x_val_and_test, y_train, y_val_and_test = train_test_split(x, y, test_size=0.2)

print ("x_train shape = ", x_train.shape)
print ("x_validate and test shape = ", x_val_and_test.shape)
```

```python
#Splitting the (testing and validation) dataset into testing and validation data. 50% as testing data and 50% as validation data. 

x_val, x_test, y_val, y_test = train_test_split(x_val_and_test, y_val_and_test , test_size=0.5)

print ("x_validate shape = ", x_val.shape)
print ("x_test shape = ", x_test.shape)
```

```python
#Defining the keras model
#Hidden Layer 01 : 512 neurons , RELU function
#Hidden Layer 02 : 512 neurons , RELU function
#Output Layer : 1 neuron , sigmoid function
#Input Shape : 20, since 20 input featuers are available

from keras.models import Sequential
from keras.layers import Dense

model = Sequential([
    Dense(256, activation='relu', input_shape=(20,)),
    Dense(256, activation='relu'),
    Dense(4, activation='softmax'),
])
model.summary()
```

```python
#Compiling the model
from tensorflow.keras.optimizers import Adam

model.compile(optimizer='Adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
#Fitting the model 

BATCH_SIZE = 100
EPOCHES = 100
VERBOSE = 1
VALIDATION_SPLIT = 0.2 

model_history = model.fit(x_train,y_train, batch_size=BATCH_SIZE,epochs=EPOCHES, verbose=VERBOSE)
```

```python
#Evaluate the model 
#Evaluating Method : Accuracy

model_accuracy = model.evaluate(x_test, y_test)[1]
print('Model accuracy:', model_accuracy)
#print("Accuracy on training set: {:.2%}".format(model.score(X_train, y_train)))
#print("Accuracy on test set: {:.2%}".format(model.score(X_test, y_test)))
```

```python
#Evaluate the model 
#Evaluating Method : Precision, Recall and F-1 score

y_probs = model.predict(x_test)
y_pred  = np.argmax(y_probs, axis=-1)

print(metrics.classification_report(y_test, y_pred))
```
